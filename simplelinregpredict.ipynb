```python
dataset = pd.read_csv('perpusdacikinikuningan2015-2019.csv')
X = dataset.iloc[:, [0, 2]].values
X_df = dataset[["tahun",  "jumlah"]]
X_df.head(10)
```

```python
# Menggunakan metode elbow untuk menentukan angka cluster yang tepat
from sklearn.cluster import KMeans
X = X_df.loc[:,["tahun", "jumlah"]]
wcss = []
k = range(1,11)
for i in k:
    means_k = KMeans(n_clusters=i, random_state=42)
    means_k.fit(X)
    wcss.append(means_k.inertia_)
plt.figure(figsize=(12,6))
plt.plot(k , wcss , 'bo-')
plt.title('Elbow Method')
plt.xlabel('Cluster Number')
plt.ylabel('WCSS')
plt.show()
```

```python
# Menjalankan K-Means Clustering ke dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
labels = kmeans.fit_predict(X_df)
X_df = X.copy()
X_df["Cluster"] = labels
#Memilih dua variabel dan visualisasikan titik datanya 
X = X_df[["tahun", "jumlah"]]
plt.scatter(X["tahun"],X["jumlah"],c='black')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```

```python
#Menentukan jumlah clusters (K) dan pilih centroid secara acak untuk setiap cluster 

K=3

#Pilih pengamatan acak sebagai centroid 
Centroids = (X_df.sample(n=K))
plt.scatter(X_df["tahun"],X_df["jumlah"],c='black')
plt.scatter(Centroids["tahun"],Centroids["jumlah"],c='red')
plt.title('Consumers Cluster')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengunjung')
plt.show()
Centroids[:]
```

Langkah berikutnya adalah menetapkan semua titik ke centroid cluster terdekat dan Hitung ulang centroid dari cluster yang baru terbentuk (dilakukan secara berulang)

```python
diff = 1
j=0

while(diff!=0):
    XD=X_df
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["tahun"]-row_d["tahun"])**2
            d2=(row_c["jumlah"]-row_d["jumlah"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X_df[i]=ED
        i=i+1

    C=[]
    for index,row in X_df.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X_df["Cluster"]=C
    Centroids_new = X_df.groupby(["Cluster"]).mean()[["jumlah","tahun"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['jumlah'] - Centroids['jumlah']).sum() + (Centroids_new['tahun'] - Centroids['tahun']).sum()
        print(diff.sum())
    Centroids = X_df.groupby(["Cluster"]).mean()[["jumlah","tahun"]]
```

Nilai-nilai ini mungkin berbeda setiap kali dijalankan. Training akan berhenti ketika centroid tidak berubah setelah dua iterasi. Awalnya didefinisikan diff = 1, kemudian didalam badan perulangan (while loop), dihitung perbedaan antara centroid pada iterasi sebelumnya dan iterasi saat ini. Training data akan berhenti ketika perbedaan sudah 0. 

```python
#Visualiasasi cluster yang terbentuk
color=['dodgerblue','magenta', 'lime']
for k in range(K):
    data=X_df[X_df["Cluster"]==k+1]
    plt.scatter(data["tahun"],data["jumlah"],c=color[k])
plt.scatter(Centroids["tahun"],Centroids["jumlah"],c='red')
plt.title('Cluster Jumlah Pengunjung Perpustakaan Daerah DKI Jakarta (2015-2019)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```

Dari hasil diagram diatas, dapat dengan jelas memvisualisasikan 3 kelompok. Titik-titik merah mewakili pusat centroid dari setiap cluster. sedangkan titik hijau muda (lime), merah keunguan (magenta) dan biru muda (dodger blue) adalah representasi dari 3 cluster yang terbentuk

# Probability Sampling
Dengan metode probability sampling,  diambil sampel 2 perpustakaan umum daerah, yakni Perpustakaan Daerah Cikini dan Kuningan sebagai data sampel. Teknik yang digunakan dalam penentuan 2 sampel tersebut adalah Simple random sampling dengan alasan bahwa Teknik tersebut dipakai untuk mendapatkan sampel yang langsung dilakukan pada unit sampling. Dimana tiap unit sampling sebagai unsur populasi memperoleh peluang yang sama untuk menjadi sampel atau untuk mewakili populasinya. Hal ini dirasa dapat dilakukan karena anggota populasi adalah sama-sama perpustakaan (Homogen).

# Outlier
Metode binning membagi kumpulan data kedalam beberapa partisi atau bin. Dimulai dengan mengurutkan setiap nilai pada sebuah atribut. Kemudian data yang sudah diurutkan dibagi kedalam beberapa partisi atau bin yang memiliki frekuensi yang sama(equal-frequency partitioning). Outlier dataset perpustakaan akan digunakan interval range untuk binning data:
- range (Rendah, Sedang, Tinggi) 
- cut & qcut

```python
import numpy as np
df = pd.read_csv('perpusdacikinikuningan2015-2019.csv')
bins = np.linspace(min(df['jumlah']), max(df['jumlah']), 4)
bins
```

```python
df = pd.read_csv('perpusdacikinikuningan2015-2019.csv')
interval_range = pd.interval_range(start=0, freq=5000, end=150000) 
kategori = ['Rendah', 'Standar', 'Tinggi']
df['Jumlah Binned'] = pd.cut(df['jumlah'], bins, labels=kategori, include_lowest=True)
df['Jumlah Binned 2'] = pd.cut(df['jumlah'], bins = interval_range)
df['Jumlah Binned 3'] = pd.qcut(df['jumlah'], 3)
df
```

# Normalisasi
1. Metode normalisasi data pertama yang digunakan untuk dataset pengunjung perpustakaan ialah simple feature scaling, yaitu dengan cara membagi setiap nilai dengan nilai maksimum pada kolom jumlah pengunjugn perpustakaan dari dataset "perpusdacikinikuningan2015-2019.csv". 
2. Metode normalisasi data kedua yang digunakan untuk dataset pengunjung perpustakaan adalah Min-Max. Cara kerjanya setiap nilai pada sebuah kolom dikurangi dengan nilai minimum kolom tersebut, kemudian dibagi dengan rentang nilai atau nilai maksimum dikurangi nilai minimum dari fitur tersebut.
3. Metode normalisasi data ketiga yang digunakan untuk dataset pengunjung perpustakaan adalah Z-score atau disebut juga standard score. Dimana masing-masing nilai pada kolom jumlah dikurangi dengan (µ) yang merupakan nilai rata-rata (mean) kolom jumlah, kemudian dibagi dengan (σ) yang merupakan standar deviasi

```python
#1
df["jumlah"] = df["jumlah"] / df["jumlah"].max()
print(df)
#2
df["jumlah"] = (df["jumlah"] - df["jumlah"].min()) / (df["jumlah"].max() - df["jumlah"].min())
print(df)
#3
df["jumlah"] = (df["jumlah"] - df["jumlah"].mean()) / df["jumlah"].std()
print(df)
```

# Feature Engineering: Derived Mathematics Features
Jenis Feature Engineering berikutnya yang digunakan untuk dataset prediksi jumlah pengunjung perpustakaan di DKI Jakarta ini adalah Derived Mathematics Features. Untuk hal tersebut, karena data yang digunakan untuk prediksi hanya 10 data pengunjung dari 2 perpustakaan, maka dapat dibentuk dalam array dengan penetapan sumbu koordinat x = Perpusda Cikini (c) dan koordinat y = Perpusda Kuningan (k), selanjutnya digunakan metode Regresi Linear Sederhana (Simple Linear Reggression)

```python
c = np.array([72758,103341,83619,79114,62895])
k = np.array([13529, 68654,53821,5447,45230])
plt.xlabel("Cikini")
plt.ylabel("Kuningan")
plt.title("Jumlah Pengunjung Perpustakaan Daerah di DKI Jakarta Tahun 2015-2019")
plt.scatter(c, k);
```

```python
from sklearn.linear_model import LinearRegression
X = c[:, np.newaxis]
model = LinearRegression().fit(X, k)
yfit = model.predict(X)
plt.xlabel("Cikini")
plt.ylabel("Kuningan")
plt.title("Jumlah Pengunjung Perpustakaan Daerah di DKI Jakarta Tahun 2015-2019")
plt.scatter(c, k)
plt.plot(c, yfit);
```

- Dari model scatterplot diatas, dapat dilihat bahwa sedikit menyulitkan untuk membaca adanya hubungan untuk kedua koordinat (sama-sama menunjukkan jumlah pengunjung), sebab meskipun sudah coba disesuaikan antara garis ke data menggunakan Linear Regression untuk mendapatkan hasil yang optimal dari hubungan diantara kedua variabel, tetap dibutuhkan suatu model feature engineering yang dapat lebih menggambarkan hubungan antara variabel c dan k.
- Salah satu pendekatan yang dapat dilakukan untuk menggambarkan hubungan tersebut adalah mengubah data dan menambahkan kolom fitur tambahan untuk menghasilkan fleksibilitas pada model yang dihasilkan. Sebagai contoh, disini digunakan fitur polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, include_bias=False)
X2 = poly.fit_transform(X)
print(X2)
```

Matriks derived features diatas memiliki satu kolom yang mewakili x, dan kolom kedua yang mewakili x2, dan kolom ketiga yang mewakili x3. Menghitung regresi linier pada input yang diperluas ini memberikan gambaran tentang hubungan data pengunjung perpustakaan Cikini (c) dan Kuningan (k)
